### Renaming Columns & changing

```python
customer_df.columns = customer_df.columns.str.lower()
customer_df.columns = customer_df.columns.str.replace(" ", "_")
customer_df
```

```python
customer_df['effective_to_date'] = customer_df['effective_to_date'].astype('datetime64[ns]')
```

### Splitting Categorial and Numerical
##### After splitting the data, I realise it is correctly split, and the dates are put into categorical, how I want it. I can now change the dates to datetime format

```python
customer_df_NUM = customer_df.select_dtypes(include = "number")
customer_df_CAT = customer_df.select_dtypes(exclude = "number")
```

```python
customer_df_CAT
```

```python
customer_df_NUM
```

### Plotting Correlation Matrix

After putting in the correlation heatmap, I realise a lot of the numerical variables are not correlated. I will keep all the columns for now


```python
correlation_matrix_num = customer_df_NUM.corr().apply(lambda x: round(x,2))
correlation_matrix_num
```

```python
correlation_matrix_num = customer_df_NUM.corr().apply(lambda x: round(x,2))
sns.heatmap(correlation_matrix_num, cmap = "coolwarm", annot=True)
plt.show
```

### Plotting the distribution Plots for continuos variables

```python
for i in customer_df_NUM.columns:
    sns.distplot(customer_df_NUM[i])
    plt.show()
```

### Why are we deleting customer_id
We need to delete it because it is a unique value for each row that is completely independent to both the target variable and the other columns. Therefore it is just filling space and confusing the model. We should delete it

```python
customer_df_CAT = customer_df_CAT.drop(columns = "customer")
```

```python
customer_df_CAT
```

For categorial variables I will use a countplot

```python
for i in customer_df_CAT.columns:
    sns.countplot(x=i, data=customer_df_CAT)
    plt.show()
```
